```python
import os
import pandas as pd
import numpy as np
```

```python
# Set the path to the 'data' folder
folder_path = './data/dataset'

# List all CSV files in the folder
all_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Create an empty list to store DataFrames
df_list = []

# Loop through the CSV files and append them to the list
for file in all_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    df_list.append(df)

# Concatenate all DataFrames
final_df = pd.concat(df_list, ignore_index=True)
```

```python
agg_df = final_df.groupby('class',as_index=False)['filename'].agg('count').rename(columns={'filename':'count'})
agg_df['tot'] = len(final_df)
agg_df['pct'] = agg_df['count']/agg_df['tot']
agg_df = agg_df.drop(columns=['tot']).sort_values('pct',ascending=False).reset_index(drop=True)
agg_df.head(50)
```
