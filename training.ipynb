```python
import os
import dotenv
import pandas as pd
import numpy as np
from tqdm import tqdm
import umap
import hdbscan
import pickle
import torch
from ultralytics import YOLO

print("GPU is available:", torch.cuda.is_available())

if torch.cuda.is_available():
  print("Current device:", torch.cuda.get_device_name(0))

device = 'cuda' if torch.cuda.is_available() else 'cpu'
yolo_device = 0 if torch.cuda.is_available() else 'cpu'

cwd = os.getcwd()
data_path = os.path.join(cwd,'data')
data_yaml_path = os.path.join(data_path,'mydata.yaml')
```

```python
model = YOLO("milplanes.pt")

results = model.train(data=data_yaml_path, epochs=100, imgsz=640, device=yolo_device)
```

```python
file_path = 'metrics_final.txt'
with open(file_path, 'r') as f:
    data = f.read()

data = [i for i in data.split(' ') if i != '']


data_dict = {'Class':[],
             'Images':[],
             'Instances':[],
             'Precision':[],
             'Recall':[],
             'mAP50':[],
             'mAP50_95':[]}

for i in range(65):
        for j,k in enumerate(list(data_dict.keys())):
            data_dict[k] = data_dict[k] + [data[j+7*i]]


data_df = pd.DataFrame(data_dict).replace('\n','',regex=True)
cols = [col for col in data_df.columns if 'Class' not in col]

data_df[cols] = data_df[cols].apply(pd.to_numeric, errors='coerce')
```
